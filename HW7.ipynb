# 利用迴圈方式，在固定神經元層數、Learning Rate、Activation、optimizer和loss的情況下，找出能夠使精確率最高的參數(神經元數)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)= mnist.load_data()
```

```python
x_train = x_train.reshape(60000,784)
x_test = x_test.reshape(10000,784)
```

```python
from keras.utils import np_utils
```

```python
y_train = np_utils.to_categorical(y_train,10)
y_test = np_utils.to_categorical(y_test,10)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation
from keras.optimizers import SGD,Adam
from ipywidgets import interact_manual
```

```python
def test(測試編號):
    plt.imshow(x_test[測試編號].reshape(28,28),cmap='Greys')
    print('神經網路判斷為:',predict[測試編號])
```

首先透過上週老師的PPT教學以及網路上的基本介紹，決定嘗試的配置為:Activation用relu，loss用categorical_crossentropy，,optimizer用Adam，而神經元學習的層數則先固定為兩層。

再來設定要跑迴圈的參數，i代表第一層的神經元數，j代表第二層的神經元數。將神經元數限縮在10-20，而i、j的迴圈間隔皆為1(範圍限縮得有點小是因為測試後發現電腦有點跑不動)

```python
besti =0 
bestj =0
bestcorr=0 #用來存放表現最好時的i j k以及精確率
for i in range(10,20):
    for j in range(10,20):
        print(str(i)+','+str(j))
        model = Sequential()
        model.add(Dense(i,input_dim=784))
        model.add(Activation('relu'))
        model.add(Dense(j))
        model.add(Activation('relu'))
        model.add(Dense(10))
        model.add(Activation('softmax'))
        model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.001),metrics =['accuracy'])
        model.fit(x_train,y_train,batch_size=100,epochs=20)
        score = model.evaluate(x_test,y_test)
        if score[1]>bestcorr: #當此次配置的精確率比之前最高的精確率高時，則更新最佳為此次的係數，並將此次模型存取
            bestcorr = score[1]
            besti = i
            bestj = j
        if score[1]>0.9:
            model_json= model.to_json()
            open('i'+str(i)+'j'+str(j)+'.json','w').write(model_json)
            model.save_weights('i'+str(i)+'j'+str(j)+'.h5')
print("第一層最佳神經元數:"+str(besti)) #迴圈結束後，印出最佳的結果以及參數
print("第二層最佳神經元數:"+str(bestj))  
print("最佳正確率:"+str(bestcorr))
```

得到的結果為:
第一層最佳神經元數:18
第二層最佳神經元數:14
最佳正確率:0.9477
