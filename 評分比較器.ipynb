# 單純就文字的出現評分

開友達字詞

```python
import csv
AUO_words = []
# 開啟 CSV 檔案
with open('sort_freqAUO.csv','r',encoding='UTF-8-sig') as csvf:
    rows = csv.reader(csvf)
    for row in rows:
        AUO_words.append(row)
csvf.close()
```

```python
AUO_words
```

```python
only_words = []
for cell in AUO_words:
    only_words.append(cell[0])
```

```python
only_words
```

開標竿字詞

```python
import csv
Leader_words = []
# 開啟 CSV 檔案
with open('final_keywords.csv','r',encoding='UTF-8-sig') as csvf:
    rows = csv.reader(csvf)
    for row in rows:
        Leader_words.append(row)
csvf.close()
```

```python
for token in Leader_words:
    while token[-1] == '':
        del(token[-1]) #去除一個尾巴的空格
```

```python
Leader_words
```

先計算總數

```python
E_total = 0
S_total = 0
G_total = 0
N_total = 0
for cell in Leader_words:
    if cell[0] =='E':
        E_total+=1
    if cell[0] =='S':
        S_total+=1 
    if cell[0] =='G':
        G_total+=1
    if cell[0] =='N':
        N_total+=1
dic_total = {'E_total':E_total,'S_total':S_total,"G_total":G_total,"N_total":N_total}
dic = {'E':0,'S':0,"G":0,"N":0}
```

```python
dic_total
```

```python
dic
```

開始搜尋

```python
missing_words = []
for cell in Leader_words:
    switch1 = 0
    for word in cell[2:]:
        if word in only_words:
            dic[str(cell[0])] +=1
            switch1 = 1
            break
    if switch1 == 0:
        missing_words.append(cell)
    
```

```python
import matplotlib.pyplot as plt
```

```python
def createLabels(data):                   # 自定義函數，為引用plusone團隊
    for item in data:
        height = item.get_height()
        plt.text(
            item.get_x()+item.get_width()/2., 
            height*1.05, 
            '%d' % int(height),
            ha = "center",
            va = "bottom",
        )
```

```python
width = 0.4
A = plt.bar([i-width/2 for i in range(len(dic))],dic.values(), width=width,label='AUO')
B = plt.bar([i+width/2 for i in range(len(dic))], dic_total.values(), width=width,label='Leader')
plt.xticks(range(len(dic)), dic.keys())

createLabels(A)
createLabels(B)

plt.legend()
plt.show()
plt.savefig("Compare_bin.png")   
```
