```python
import csv
read_tokens = []
# 開啟 CSV 檔案
with open('processed_contentLeader.csv','r',newline='',encoding='UTF-8-sig') as csvf:
    rows = csv.reader(csvf)
    for row in rows:
        read_tokens.append(row)
csvf.close()
```

```python
for token in read_tokens:
    if len(token) == 0:
        continue
    elif token[-1] == '':
        del(token[-1]) #去除一個尾巴的空格
read_tokens[10]
```

```python
def super_context(tokens,keyword,length):
    context = []
    for i in range(len(tokens)):
        for word in tokens[i]:
            if word == keyword:
                context.append(tokens[i-length:i+length+1])
                break
    return context
                
                
        
    
```

```python
searchword = "tcfd"
```

```python
a = super_context(read_tokens,searchword,10)
```

```python
a[1]
```

```python
imerge_token=[]
for cell in a:
    for sent in cell:
        for word in sent:
            imerge_token.append(word)
```

```python
imerge_token[1:100]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(imerge_token)
for key,val in freq.items():
    print (str(key) + ':' + str(val))
```

```python
sort = sorted(freq.items(), key=lambda x: x[1],reverse = True)
```

```python
print(sort[:20])
```

```python
keywords = []
# 開啟 CSV 檔案
with open('final_keywords.csv','r',encoding='UTF-8-sig') as csvf:
    rows = csv.reader(csvf)
    for row in rows:
        keywords.append(row)
csvf.close()
```

```python
for token in keywords:
    while token[-1] == '':
        del(token[-1]) #去除一個尾巴的空格
```

```python
keywords
```

```python
searchword
```

```python
wordbag = {}
for cell in sort :
    switch = 0
    for line in keywords:           
        if cell[0] in line and cell[0] != searchword:
            switch = 1
            break
    if switch == 1:
        wordbag.setdefault(cell[0],cell[1])
        
```

多重關鍵字畫關聯圖(需要執行完上面的東西再跑)

```python
searchlist = ['climate_change','cdp','employee_care','csr']
```

```python
nodes = []
edges = []
widths = []
name_size = []
for bigword in searchlist:
    a = super_context(read_tokens,bigword,10)
    imerge_token=[]
    for cell in a:
        for sent in cell:
            for word in sent:
                imerge_token.append(word)
    
    freq = nltk.FreqDist(imerge_token)
    sort = sorted(freq.items(), key=lambda x: x[1],reverse = True)
    
    wordbag = {}
    for cell in sort :
        switch = 0
        for line in keywords:           
            if cell[0] in line and cell[0] != bigword:
                switch = 1
                break
        if switch == 1:
            wordbag.setdefault(cell[0],cell[1])
            
    sort1 = sorted(wordbag.items(), key=lambda x: x[1],reverse = True)
    for cell in sort1[:10]:
        nodes.append(cell[0])
        edges.append((bigword,cell[0]))
        widths.append(int(cell[1])/50)    
    nodes.append(bigword)    

    freqs = []
    # 開啟 CSV 檔案
    with open('sort_freqLeader.csv','r',newline='',encoding='UTF-8-sig') as csvf:
        rows1 = csv.reader(csvf)
        for row in rows1:
            freqs.append(row)
    csvf.close()
    
    sizes = []
    for words in nodes:
        for cell in freqs:
            if words == cell[0]:
                sizes.append((words,int(cell[1])*10)) #現在把字詞也丟進來
                break
    name_size += sorted(set(sizes), key = sizes.index)
    
    
DG1 = nx.MultiDiGraph()

DG1.add_nodes_from([w[0] for w in name_size])
DG1.add_edges_from(edges)



plt.figure(figsize=(40,40))
nx.draw_spring(DG1,with_labels=True,node_size=[w[1] for w in name_size],node_color = 'green', width=widths)
plt.show() 
```
