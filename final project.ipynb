接著取出t-5到t-1天的溫度作為輸入資料(x)，第t日的資料為輸出(y)
*資料要扣除標題、前五個元素(因為會造成輸入值不足五個)、以及最後一個元素(因為沒有輸出))

```python
x = []
y = []
for i in range(len(d_arr)-7): #扣掉標題，以及前面五個數據，還有最後一個數據
    element = [float(d_arr[i+1][5]),float(d_arr[i+2][5]),float(d_arr[i+3][5]),float(d_arr[i+4][5]),float(d_arr[i+5][5])]
    x.append(element)
    y.append([float(d_arr[i+6][5])])
```

將x、y轉換為矩陣

```python
X = np.array(x)
```

```python
Y = np.array(y)
```

嘗試繪製出模式預測的溫度資料以及測試資料的實際溫度來分析

```python
def plot_results(predicted_data, true_data):
    fig = plt.figure(facecolor='white')
    ax = fig.add_subplot(111)
    ax.plot(true_data, label='True Data')
    plt.plot(predicted_data, label='Prediction')
    plt.legend()
    plt.show()
```

```python
plot_results(predicted,y_test)
```

```python
plot_results(predicted[0:100],y_test[0:100])
```
